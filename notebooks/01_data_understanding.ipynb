```python
def top_teams_by_wins(matches):
    """
    Return teams sorted by total wins.
    """

    return (
        matches["winner"]
        .value_counts()
        .reset_index()
        .rename(columns={
            "index":"Team",
            "winner":"Wins"
        })
    )
```

```python
top_teams_by_wins(matches)
```

```python
def toss_analysis(matches):
    """
    Analyze toss winners and decisions.
    """

    return pd.crosstab(
        matches["toss_decision"],
        matches["winner"] == matches["toss_winner"]
    )
```

```python
toss_analysis(matches)
```
